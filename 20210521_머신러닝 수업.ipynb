```python
def plot_(X_resampled,y_resampled,remove=True):
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    
    if remove == True:
        
        idx_samples_remove = np.setdiff1d(np.arange(X.shape[0]),idx_resampled)
        
        idx_class_0 = y_resampled ==0
        
        plt.scatter(X_resampled[idx_class_0,0],X_resampled[idx_class_0,1],alpha=.5,label='class0')
        plt.scatter(X_resampled[~idx_class_0,0],X_resampled[~idx_class_0,1],alpha=.5,label='class1')
        plt.scatter(X[idx_samples_remove,0],X[idx_samples_remove,1],alpha=.5,label='removed')
        
    else:
        
        idx_class_0 = y_resampled ==0
        
        plt.scatter(X_resampled[idx_class_0,0],X_resampled[idx_class_0,1],alpha=.5,label='class0')
        plt.scatter(X_resampled[~idx_class_0,0],X_resampled[~idx_class_0,1],alpha=.5,label='class1')
        
    
    plt.legend()
    plt.tight_layout()
    plt.show()
    
```

```python
plot_(X, Y, remove = False)
```

```python
rus = RandomUnderSampler()

X_resampled, Y_resampled = rus.fit_resample(X,Y)
```

```python
idx_resampled = rus.sample_indices_
```

```python
plot_(X_resampled, Y_resampled, remove = False)
plot_(X_resampled, Y_resampled, remove = True)
```
